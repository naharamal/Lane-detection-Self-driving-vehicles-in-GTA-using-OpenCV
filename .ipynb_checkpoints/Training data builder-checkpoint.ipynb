```python
import win32api as wapi
keylist=["\b"]
keys="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for key in keys:
    keylist.append(key)
def getkey():
    output=[]
    for key in keylist:
        if wapi.GetAsyncKeyState(ord(key)):
            output.append(key)
    return output
```

```python
import time
last=time.time()
while True:
    
    letter=getkey()
    print(letter,end='')
    if time.time()-last>5:
        break
```
